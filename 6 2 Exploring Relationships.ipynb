## QN6
Create a categorical plot and interpret the results.

### trip duration

```python
sns.histplot(df['trip_duration'], bins = 20, kde = True)
```

```python
df["trip_duration"].describe()
```

```python
#new flag colomn depending on the duration
df.loc[df['trip_duration'] < 395, 'duration_flag'] = 'short_dur'
df.loc[(df['trip_duration'] >=395) & (df["trip_duration"] <=999), 'duration_flag'] = 'average_dur'
df.loc[df['trip_duration'] > 999, 'duration_flag'] = 'long_dur'
df.loc[df['trip_duration'] >2500, 'duration_flag'] = 'v_long_dur'
```

```python
df["duration_flag"].value_counts()
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the duration categories created above and start hour

sns.set(style="ticks")
g = sns.catplot(x="start_hour", y="trip_duration", hue="duration_flag", data=df)
```

As we can see the long/vlong trip are clearly less between 1 and 6 am, a possible reason is that long trips are used for work or school trips.

```python
# Create a categorical plot in seaborn using the duration categories created above and weekday

sns.set(style="ticks")
g = sns.catplot(x="weekday", y="trip_duration", hue="duration_flag", data=df)
```

aggerating the duration trip by week of days, it is harder to see a difference. Sometimes it's better to drill down to query the differences.
